### Function to show image using opencv

```python
'''function to read and show image from its path'''
def image_show(image_path):
    img = cv2.imread(image_path)
    cv2.imshow(image_path.split(".")[0],img)
    key = cv2.waitKey(0)
    if key == ord("b"):
        cv2.destroyAllWindows()

image_show("download.jpeg")
```

### Image me color extraction 

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
a = np.ones((10,30,3),dtype="uint8")
a[:,:10,:]=[255,0,0]
a[:,10:20,:]=[0,255,0]
a[:,20:,:]=[0,0,255]
plt.imsave("arr.png",a)
b = cv2.cvtColor(a,cv2.COLOR_RGB2GRAY)
plt.imshow(b,cmap="gray")
```

```python
import skimage as ski

img = ski.data.astronaut()
plt.imshow(img)
```

```python
img_b = img[:,:,-1]
plt.imshow(img_b,cmap="Blues")
```

```python
img_g = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
sub = cv2.subtract(img_b,img_g)                      
plt.imshow(sub, cmap="gray")
```

```python
th , img_t = cv2.threshold(sub, 60 , 255 , cv2.THRESH_BINARY)
plt.imshow(img_t,cmap="gray")
```

```python
img_clean = ski.morphology.remove_small_objects(img_t.astype(bool),50)       # morphology function is used to filtering image
plt.imshow(img_clean,cmap=("gray"))
```
