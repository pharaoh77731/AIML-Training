```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
plus = np.zeros((255, 255), dtype= int)
plus[:,50:100] = 255
plus[50:100,:] = 255
plt.imshow(plus)
```

```python
filter = np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]])
```

```python
def cov_2d(img,kernal):
    
    i_dim, j_dim = img.shape
    k_dim = kernal.shape[0]
    o_dim = i_dim - k_dim + 1

    output = np.zeros((o_dim, o_dim), dtype=int)

    for i in range(o_dim):
        for j in range(o_dim):

            region = img[i:i+k_dim,j:j+k_dim]
            output[i,j] = np.sum(region * kernal)

    return output

cov_image = cov_2d(plus, filter)
print(cov_image)
plt.imshow(cov_image)
```
